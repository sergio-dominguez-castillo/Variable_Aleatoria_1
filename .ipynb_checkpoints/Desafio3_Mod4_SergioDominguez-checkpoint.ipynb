5. Construye un modelo de distribución binomial que modele el experimento
correspondiente a escoger 5 madres al azar y observar cuántas fuman. Utilizando
este modelo y los datos reales, genera en cada caso 1000 muestras de tamaño 8 con
sus respectivas medias muestrales. A partir de ello:
● compáralas gráficamente
● verifica el teorema del límite central, considerando las muestras de datos reales y el
modelo normal derivado del modelo binomial.

```python
#array con la conversion a 1 y 0 de los true y false de columna maternal.smoker
maternal_smoker = np.where(df_Baby['Maternal.Smoker'], 1, 0)

# calculo la probabilidad de columna "Maternal.Smoker" sum(true)/toal de registros de Maternal.Smoker
probabilidad_exito=df_Baby['Maternal.Smoker'].sum()/df_Baby['Maternal.Smoker'].shape[0]

#Genero las muestras "teoricas" con la distribución binomial, y almacenamos las medias muestrales
# generacion de datos 
medias_teorica=[]
for j in range(1000):
    medias_teorica.append(np.mean(np.random.binomial(5,probabilidad_exito,size=8)))

# muestras aleatorias de datos reales
medias_real=[]
for k in range(1000):
    muestra_aleatoria=[]
    
    for t in range(8):
        muestra_aleatoria.append(sum(np.random.choice(maternal_smoker,size=5,replace=False)))
        
    medias_real.append(np.mean(muestra_aleatoria))

```

```python
# graficamos todo en uno
plt.hist(medias_teorica, bins=20,density=True,alpha=0.5, color='red', label='Binomial')
sns.histplot(medias_teorica, bins=20, color='red', kde=True )
plt.hist(medias_real, bins=20,density=True,alpha=0.5,color='blue', label='Real')
sns.histplot(medias_real, bins=20, color='blue', kde=True )
plt.axvline(np.mean(medias_real), color='blue', linestyle='--', lw=2, label='media real')
plt.axvline(np.mean(medias_teorica), color='red', linestyle='--', lw=2, label='media teorica')
plt.xlabel('Media muestral')
plt.ylabel('Frecuencia')
plt.title('Medias teorica (binomial) y real')
plt.legend()
plt.show()
```

```python
# ahora en forma independiente, solamente para compararlo visualmente
# grafica de medias teoricas
plt.hist(medias_teorica, bins=20,density=True, label='Binomial')
sns.histplot(medias_teorica, bins=20, kde=True )
plt.axvline(x=np.mean(medias_teorica), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.xlabel('Peso (en onzas)')
plt.ylabel('Frecuencia')
plt.title('Medias teorica (binomial)')
plt.legend()
plt.show()
# grafica de medias real
plt.hist(medias_real, bins=20,density=True, color='blue',label='Real')
sns.histplot(medias_real, bins=20,color='blue', kde=True )
plt.axvline(x=np.mean(medias_real), color='blue', linestyle='dashed', linewidth=2, label='Media')
plt.xlabel('Peso (en onzas)')
plt.ylabel('Frecuencia')
plt.title('Medias real - muestra aleatoria')
plt.legend()
plt.show()
```

```python
# ahora la normalizamos (la teorica)
mu=np.mean(medias_teorica)
sigma=np.std(medias_teorica)
minimo=np.min(medias_teorica)
maximo=np.max(medias_teorica)

x = np.linspace(minimo, maximo,num=100)

pdf = norm.pdf(x, loc=mu, scale=sigma)


plt.hist(medias_teorica, bins=20,density=True)

plt.plot(x, pdf, color='red', label='PDF') #Graficamos la distribución normal
plt.xlabel('Medias muestrales')
plt.ylabel('Densidad de probabilidad normalizada')
plt.title('probabilidad Teorica Normalizada')
plt.legend()
plt.show()
```

# comentarios
con los graficos anteriores podemos deducir que al generar muestras con el modelo binomial y comparar la curva, esta se aproxima con mucha similitud a la normalizada, lo que demuestra que en la medida que aumentamos la cantidad de la muestra los datos tienden a una distribucion normal.
